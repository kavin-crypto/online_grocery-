```python
# Define a function to plot the pie chart with specified colors
def plot_delivery_usage_pie_chart_custom(delivery_usage_counts):
   
    # Define the custom color palette
    colors = color=['lightblue', 'lightgreen', 'lightcoral','orange']
    
    # Explode the COVID-19 pandemic slice slightly for emphasis
    explode = [0.05 if label == 'During COVID-19 pandemic' else 0 for label in delivery_usage_counts.index]
    
    # Create a figure with subplots to allow legend on the right
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot the pie chart on the left side without labels (use only percentages)
    wedges, texts, autotexts = ax.pie(
        delivery_usage_counts,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        explode=explode,
        wedgeprops={'edgecolor': 'white', 'linewidth': 1}  # White edge for contrast
    )
    
    # Customize percentage text style
    for autotext in autotexts:
        autotext.set_color('white')  
        autotext.set_fontsize(12)
        autotext.set_weight('bold')
    
    # Add the legend on the right side
    ax.legend(wedges, delivery_usage_counts.index, loc="center left", bbox_to_anchor=(1, 0.5), fontsize=12)
    
    # Set the title and layout
    plt.title("Trends in Adoption of Online Grocery Delivery Services - Pandamic Prespective", fontsize=14, fontweight='bold')
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Call the function to plot the chart
plot_delivery_usage_pie_chart_custom(delivery_usage_counts)
```

```python

columns_of_interest = [
    "How often did you use online grocery delivery services before the COVID-19 pandemic?",
    "How often did you use online grocery delivery services during the COVID-19 pandemic?",
    "How often do you use online grocery delivery services now?"
]
frequency_data = data[columns_of_interest]

# Map frequency responses to numeric values
frequency_mapping = {
    'Never': 1,
    'Rarely': 2,
    'Sometimes': 3,
    'Often': 4,
    'Always': 5
}

# Replace string values with their corresponding numeric values
frequency_data = frequency_data.replace(frequency_mapping)

# Calculate the mean frequency of usage for each period
mean_frequency = frequency_data.mean()

# Prepare data for plotting
categories = ['Before COVID-19', 'During COVID-19', 'After COVID-19']
values = mean_frequency.values

# Create the bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=['lightblue', 'lightgreen', 'lightcoral'])

# Remove grid lines
plt.grid(False)

# Add labels and title
plt.title("Shift in Frequency of Usage of Online Grocery Delivery Services - Pandameic Perspective", fontsize=14, fontweight='bold')
plt.xlabel("Time Period", fontsize=12)
plt.ylabel("Frequency of Use (1-5)", fontsize=12)

# Add the numeric value on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Prepare data for plotting
reasons_before = data['What were the main reasons for using or not using online grocery delivery services before the COVID-19 pandemic? (Select all that apply)'].str.get_dummies(sep=', ')
reasons_during = data['What motivated you to use online grocery delivery services during the COVID-19 pandemic? (Select all that apply)'].str.get_dummies(sep=', ')
reasons_after = data['What factors currently motivate you to use online grocery delivery services?(Select all that apply)'].str.get_dummies(sep=', ')

# Sum up each reason across all responses
summary_before = reasons_before.sum()
summary_during = reasons_during.sum()
summary_after = reasons_after.sum()

# Create a DataFrame for plotting
summary_df = pd.DataFrame({
    'Before COVID-19': summary_before,
    'During COVID-19': summary_during,
    'After COVID-19': summary_after
})

# Plotting a stacked bar chart
ax = summary_df.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='viridis', grid=False)

# Add title and labels
plt.title('Comparative Analysis on Factors for Using Online Grocery Delivery Services - Pandemic Perspective', fontsize=16)
plt.xlabel('Reasons', fontsize=12)
plt.ylabel('Count of Responses', fontsize=12)

# Adjust x-ticks for better alignment and readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Move legend to the right
plt.legend(title='Time Period', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add values on top of each segment in the bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # Only annotate if height is greater than zero
        # Calculate the y position for the annotation based on the bottom of the segment
        y_position = p.get_y() + height / 2  # Centered vertically within the segment
        
        # Determine color based on background brightness
        color = p.get_facecolor()  # Get the color of the patch
        brightness = (color[0] * 299 + color[1] * 587 + color[2] * 114) / 1000  # Calculate brightness
        
        # Set text color based on brightness
        text_color = 'white' if brightness < 0.5 else 'black'
        
        ax.annotate(f'{height}', 
                    (p.get_x() + p.get_width() / 2., y_position),  # Centered horizontally above the segment
                    ha='center', va='center', fontsize=9, color=text_color)

plt.tight_layout()
plt.show()
```
